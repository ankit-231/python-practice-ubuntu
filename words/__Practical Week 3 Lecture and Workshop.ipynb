## Logical Operators:
* Logical operators in Python are used to combine conditional statements.



### Combining Comparison Operators with Logical Operators:
* You can combine comparison operators with logical operators like and, or, and not to form more complex boolean expressions..

* ***and:*** The and operator returns True if both expressions are True.

```python
x = 5
y = 15
result = (x > 2) and (y <= 15)
print(result)
```

* ***or:*** The or operator returns True if at least one of the expressions is True.

```python
x = 5
y = 10
result = (x > 7) or (y < 10)
print(result)
```

* ***not:*** The not operator returns True if the expression is False, and False if the expression is True.

```python
x = 5
result = not (x > 10)
print(result)
```


### Combining Logical Operators
* Logical operators can be combined to form more complex expressions:

* ***Combining and and or:*** 

```python
age = 20
has_id = True
is_member = False
can_enter_club = (age >= 18 and has_id) or is_member
print(can_enter_club)
```

* ***Combining not with and and or:*** 

```python
is_weekend = True
is_holiday = False
has_work = False
can_relax = (is_weekend or is_holiday) and not has_work
print(can_relax)
```

* ***if Statement:*** <span style="color: red;">***Conditional:***.</span> </br>
The if statement introduces two scenarios that are decided based on a condition.

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

* ***else Statement:*** <span style="color: red;">***Alternative execution:***.</span> </br>
The else statement provides an alternative block of code that is executed if none of the preceding if or elif conditions are met.

```python
x = 15
if x > 15:
    print("x is greater than 15")
else:
    print("x is not greater than 15")
```

* ***elif Statement (else if):*** <span style="color: red;">***Chained conditionals:***.</span> </br> 
The elif statement provides a way to handle multiple conditions in a chained manner, where each condition has a specific resolution.

```python
x = 20
if x > 15:
    print("x is greater than 15")
elif x == 10:
    print("x is 10")
else:
    print("x is less than 10")
```

# Recursion:

Recursion is a programming technique where a function calls itself directly or indirectly to solve a problem. The recursive function typically solves a small piece of the problem and then makes a recursive call to solve the rest of the problem.

***Key Concepts of Recursion:***

* ***1.	Base Case:*** condition under which the recursion stops.
* ***2.	Recursive Case:*** part of the function where the function calls itself to work towards the base case.

```python
factor_number = 5

def factorial(n):
    if n == 0:
        print(f"Base case: {n}")
        print(f"Recursive Case starts at: {factor_number}")
        return 1
    else:
        return n * factorial(n - 1)


print(factorial(factor_number))
```

* ***n!:*** factorial of a non-negative integer

```python
# Fibonacci Sequence
"""
    Please create a function to compute the 
    Fibonacci sequence. You must use recursion, 
    applying the explanation of recursion from 
    last week.
    @ F(0) = 0
    @ F(1) = 1
    @ F(n) = F(n-1) + F(n-2) for n > 1

"""
def fibo_seq(n):
    if n <=1:
        return n
    else:
        return fibo_seq(n-1) + fibo_seq(n-2)
```
